```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')


import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
train= pd.read_csv('C:/Users/lovej/Desktop/python/Kaggle/titanic/train.csv')

data= pd.read_csv('C:/Users/lovej/Desktop/python/Kaggle/titanic/test.csv')
```

```python
print('Oldest Passenger was of:',data['Age'].max(),'Years')
print('Youngest Passenger was of:',data['Age'].min(),'Years')
print('Average Age on the ship:',data['Age'].mean(),'Years')
```

```python
data['Initial']=0
for i in data:
    data['Initial']=data.Name.str.extract('([A-Za-z]+)\.')
    
#let's extract the Salutations(mr, mrs,,,)

train['Initial']=0
for i in train:
    train['Initial']=train.Name.str.extract('([A-Za-z]+)\.')
    
#let's extract the Salutations(mr, mrs,,,)
```

```python
pd.crosstab(data.Initial,data.Sex).T.style.background_gradient(cmap='summer_r')

# Checking the Initials with Sex
```

Okay so there are some misspelled Initials like Mlle or Mme that stand for Miss. I will replace them with Miss and same thing for other values.

```python
data['Initial'].replace(['Col','Dona','Dr','Ms','Rev'],
                        ['Other','Mrs','Mr','Miss','Other'],inplace=True)

train['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady',
                        'Countess','Jonkheer','Col','Rev','Capt','Sir',
                        'Don'],['Miss','Miss','Miss','Mr','Mr','Mrs',
                                'Mrs','Other','Other','Other','Mr','Mr','Mr'],inplace=True)
```

```python
data.groupby('Initial')['Age'].mean()
#Let's check the average age by Initials
```

Filling NaN Ages

```python
## Assigning the NaN values with the ceil values of the mean ages
data.loc[(data.Age.isnull())&(data.Initial=='Mr'),'Age']=33
data.loc[(data.Age.isnull())&(data.Initial=='Mrs'),'Age']=39
data.loc[(data.Age.isnull())&(data.Initial=='Master'),'Age']=8
data.loc[(data.Age.isnull())&(data.Initial=='Miss'),'Age']=22
data.loc[(data.Age.isnull())&(data.Initial=='Other'),'Age']=43

## Assigning the NaN values with the ceil values of the mean ages
train.loc[(train.Age.isnull())&(train.Initial=='Mr'),'Age']=33
train.loc[(train.Age.isnull())&(train.Initial=='Mrs'),'Age']=36
train.loc[(train.Age.isnull())&(train.Initial=='Master'),'Age']=5
train.loc[(train.Age.isnull())&(train.Initial=='Miss'),'Age']=22
train.loc[(train.Age.isnull())&(train.Initial=='Other'),'Age']=46
```

```python
data.Age.isnull().any()
# No null values left finally
train.Age.isnull().any()
# No null values left finally
```

```python
train['Embarked'].fillna('S',inplace=True)
```

```python
data.Embarked.isnull().any()
```

```python
print('Highest Fare was:',data['Fare'].max())
print('Lowest Fare was:',data['Fare'].min())
print('Average Fare was:',data['Fare'].mean())
```

```python
data.Age.min()
```

```python
data['Age_band']=0
data.loc[data['Age']<=15.2,'Age_band']=0
data.loc[(data['Age']>15.2)&(data['Age']<=30.4),'Age_band']=1
data.loc[(data['Age']>30.4)&(data['Age']<=45.6),'Age_band']=2
data.loc[(data['Age']>45.6)&(data['Age']<=60.8),'Age_band']=3
data.loc[data['Age']>60.8,'Age_band']=4
data.head()


train['Age_band']=0
train.loc[train['Age']<=16,'Age_band']=0
train.loc[(train['Age']>16)&(train['Age']<=32),'Age_band']=1
train.loc[(train['Age']>32)&(train['Age']<=48),'Age_band']=2
train.loc[(train['Age']>48)&(train['Age']<=64),'Age_band']=3
train.loc[train['Age']>64,'Age_band']=4
train.head()
```

```python
data['Age_band'].value_counts().to_frame().style.background_gradient(cmap='summer')
#checking the number of passengers in each band
```

```python
data['Family_Size']=0
data['Family_Size']=data['Parch']+data['SibSp']
data['Alone']=0
data.loc[data.Family_Size==0,'Alone']=1

train['Family_Size']=0
train['Family_Size']=train['Parch']+train['SibSp']
train['Alone']=0
train.loc[train.Family_Size==0,'Alone']=1
```

```python
data['Fare_Range']=pd.qcut(data['Fare'],4)
data.groupby(['Fare_Range'])['Age'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

```python
data['Fare_cat']=0
data.loc[data['Fare']<=7.896,'Fare_cat']=0
data.loc[(data['Fare']>7.896)&(data['Fare']<=14.454),'Fare_cat']=1
data.loc[(data['Fare']>14.454)&(data['Fare']<=31.5),'Fare_cat']=2
data.loc[(data['Fare']>31.5)&(data['Fare']<=512.329),'Fare_cat']=3

train['Fare_cat']=0
train.loc[train['Fare']<=7.91,'Fare_cat']=0
train.loc[(train['Fare']>7.91)&(train['Fare']<=14.454),'Fare_cat']=1
train.loc[(train['Fare']>14.454)&(train['Fare']<=31),'Fare_cat']=2
train.loc[(train['Fare']>31)&(train['Fare']<=513),'Fare_cat']=3
```

```python
data['Sex'].replace(['male','female'],[0,1],inplace=True)
data['Embarked'].replace(['S','C','Q'],[0,1,2],inplace=True)
data['Initial'].replace(['Mr','Mrs','Miss','Master','Other'],
                       [0,1,2,3,4],inplace=True)

train['Sex'].replace(['male','female'],[0,1],inplace=True)
train['Embarked'].replace(['S','C','Q'],[0,1,2],inplace=True)
train['Initial'].replace(['Mr','Mrs','Miss','Master','Other'],
                       [0,1,2,3,4],inplace=True)
```

```python
train['Fare_Range']=pd.qcut(train['Fare'],4)
train.groupby(['Fare_Range'])['Survived'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

```python
data.drop(['Name','Age','Ticket','Fare','Cabin','Fare_Range'],
         axis=1, inplace=True)
train.drop(['Name','Age','Ticket','Fare','Cabin','Fare_Range'],
         axis=1, inplace=True)
```

```python
features=['Pclass','Sex','SibSp','Parch',
          'Embarked','Initial','Age_band',
          'Family_Size','Alone','Fare_cat']
target='Survived'
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn. ensemble import AdaBoostClassifier

clf=AdaBoostClassifier(n_estimators=200,random_state=0,learning_rate=0.1)
clf.fit(train[features],train[target]) 
pred=clf.predict(data[features])
```

```python
pred
```

```python
submission = pd.DataFrame({'PassengerId':data['PassengerId'], 'Survived': pred})
submission.head()
```

```python
filename = 'Titanic Predictions 1.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
```
